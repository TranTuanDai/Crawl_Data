```python

keywords = ['Xã', 'Phường', 'Thị Trấn','xã ','Tt.','Thị trấn','thị trấn','phường','P. ','TT.']

```

```python


def extract_address_info(address):
   
    words = address.split(',')
    
    
    for word in words:
        for keyword in keywords:
            if keyword in word:
                return word.strip()  
    return None  

```

```python

extracted_df['Address'] = extracted_df['Address'].apply(extract_address_info)

```
