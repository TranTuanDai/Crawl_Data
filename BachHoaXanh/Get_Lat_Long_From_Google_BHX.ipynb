```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import time
import pandas as pd
```

```python
df=pd.read_excel('/FRT/24.Doanh_thu_BHX/Shop_Info/shop_goc_bhx.xlsx')
df
```

```python
pd.set_option('display.max_columns',40)
options = webdriver.ChromeOptions()
options.add_argument('start-maximized')
options.add_argument('disable-infobars')
options.add_argument('--disable-extensions')
options.add_argument("--disable-gpu")
options.add_argument('--incognito')
url='https://www.google.com/'
```

```python
service = Service('/ChromeDriver/chromedriver_win32/chromedriver.exe')
driver = webdriver.Chrome(service=service,options=options)
driver.get(url)
```

```python
address_column = "BHX_ADDRESS"
for index, row in df.iterrows():
    driver.get(url)
    address = row[address_column]
    search_box = driver.find_element(By.ID, 'APjFqb')
    search_box.clear()
    search_box.send_keys(address)
    button_click=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[4]/center/input[2]')
    button_click.click()
    time.sleep(1.5)
    try:
        a_tag=driver.find_element(By.XPATH,'/html/body/section/div[1]/div[1]/aside/article/div[2]/div[1]/a')
        href=a_tag.get_attribute('href')
        df.at[index, "link_address"] = href
        print(address," :", href)
    except:
        df.at[index, "link_address"] = "Không tìm thấy"
        print('Không tìm thấy')
```

```python
df['lat'] = df['link_address'].str.extract(r'query=([\d\.]+),([\d\.]+)', expand=True)[0]
df['long'] = df['link_address'].str.extract(r'query=([\d\.]+),([\d\.]+)', expand=True)[1]
```
