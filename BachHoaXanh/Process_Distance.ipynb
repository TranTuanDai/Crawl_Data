```python
from geopy.distance import geodesic
import pandas as pd
import numpy as np
import geopy.distance
```

```python
df_frt=pd.read_excel("/FRT/24.Doanh_thu_BHX/4.Bach_Hoa_Xanh/1.Latitude_Longitude/Lat_long_FLC.xlsx")
df_frt
```

```python
df_bhx=pd.read_excel('/FRT/24.Doanh_thu_BHX/4.Bach_Hoa_Xanh/1.Latitude_Longitude/Lat_long_bhx.xlsx',sheet_name='data')
df_bhx
```

```python
df_frt['SHOP_CODE']=df_frt['SHOP_CODE'].astype(str)
df_bhx['shop_code_bhx']=df_bhx['shop_code_bhx'].astype(str)
```

```python
def calculate_distance(row, point_a):
    shop_code_bhx=row['shop_code_bhx']
    point_b = (row['lat'], row['long'])
    distance = geopy.distance.geodesic(point_a, point_b).kilometers
    return shop_code_bhx,distance
```

```python
Shop_LC=[]
Shop_BHX=[]
Distances=[]
```

```python


for index,row in df_frt.iterrows():
    point_a = (row['LATITUDE'], row['LONGITUDE'])
    shop_code_frt=row['SHOP_CODE']
    for index,row in df_bhx.iterrows():
        shop_code_bhx,distance=calculate_distance(row,point_a=point_a)
        if shop_code_bhx is None:
            Shop_LC.append(shop_code_frt)
            Shop_BHX.append('Không có shop gần bán kính 2km')
            Distances.append('')
        elif distance <=5:
            Shop_LC.append(shop_code_frt)
            Shop_BHX.append(shop_code_bhx)
            Distances.append(distance)
```

```python
print(len(Shop_LC))
print(len(Shop_BHX))
print(len(Distances))
```

```python
df=pd.DataFrame()
df['Shop_code_FLC']=Shop_LC
df['Shop_code_BHX']=Shop_BHX
df['Distance']=Distances
```
