```python
def column_type_transform(df, col_dict, *cols):
    '''convert strings to datetimes'''
    df2 = df.copy()
    for col in cols:
        col_type = col_dict[col][1]
        datetime_type = col_dict[col][-1]
        try:
            # remove the apostrophe
            df2[col] = df2[col].apply(lambda x: str(x).replace("'",""))
            
            if col_type =='datetime64[ns]':
                if datetime_type == 'dmy':
                    df2[col] = pd.to_datetime(df2[col], format = '%d/%m/%Y %H:%M:%S', errors='coerce')
                elif datetime_type == 'd-m-y':
                    df2[col] = pd.to_datetime(df2[col], format = '%d-%m-%Y %H:%M:%S', errors='coerce')
                else:
                    df2[col] = pd.to_datetime(df2[col], errors='coerce')
            elif col_type == 'date':
                if datetime_type == 'dmy':
                    df2[col] = pd.to_datetime(df2[col], format = '%d/%m/%Y', errors='coerce')
                elif datetime_type == 'd-m-y':
                    df2[col] = pd.to_datetime(df2[col], format = '%d-%m-%Y', errors='coerce')
                else:
                    df2[col] = pd.to_datetime(df2[col], errors='coerce')
            else:
                df2[col] = df2[col].astype(col_type)
        except:
            df2[col] = df2[col]
    return df2
```

```python
def unique_column(cols):
    col_len = len(cols)
    result = []
    for i in range(col_len):
        col = cols[i].strip()
        col_query = [col.lower() for col in cols[:i]]
        count = col_query.count(col.lower())
        if count > 0:
            result.append(col + ' v' + str(count))
        else:
            result.append(col)
    return result
```

```python
def date_float_to_time(val):
    result = ''
    if val == val:
        minute, sec = divmod(val * 86400, 60)
        hr, minute = divmod(minute, 60)
        result = datetime.strptime("{:02.0f}:{:02.0f}:{:02.0f}".format(hr, minute, sec), '%H:%M:%S')
    else:
        result = np.nan
    
    return result

def max_char_length(df):
    for col in df.columns:
        print(col, ':', df[col].dtypes, ':', max([len(str(i))for i in df[col]]))
```
